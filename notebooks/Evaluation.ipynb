```python
import os
import torch
import pandas as pd
import sys
import glob

# set device
device = 'cuda' if torch.cuda.is_available() else 'cpu'
```

```python
# set path
root_dir = "gdrive/My Drive/Master_Thesis/"
base_dir = os.path.join(root_dir, "predictions/Base/")
sexism_dir = os.path.join(root_dir, "predictions/Sexism/")
test_dir = os.path.join(root_dir, "predictions/Small/")
```

```python
pred_dir = test_dir
```

```python
# import utilities
sys.path.append(os.path.join("/content/", root_dir))

import evaluation as eva
from utilities import get_datetime
```

```python
# read names of all prediction files
files = glob.glob(pred_dir + "*.csv")

# Read csv file into dataframe
dfs = []
infos = []

for f in files:
  df = pd.read_csv(f)
  df = df.drop("Unnamed: 0", axis=1)
  file_name = f.replace(pred_dir, "")
  attr = file_name.replace(".csv", "").split("_")

  for idx, pred in enumerate(df["Prediction"].tolist()):
    if pred != pred:
      df.loc[idx, "Prediction"] = "None"

  info_ = {"Model_Name": attr[0],
           "Model_Version": attr[1],
           "Test_Set": attr[2],
           "Prediction_File": file_name,
           }

  dfs.append(df)
  infos.append(info_)

args = {"batch_size": 15,
        "threshold": 0.95,
        "n-gram": 4,
        "device": device}
```

```python
results = eva.evaluation_pipeline(dfs, infos, args)
```

```python
save_dir = os.path.join(root_dir, 'evaluation/evaluation_') + info_["Test_Set"] + "_" + get_datetime("%d,%m,%Y--%H,%M") + '.csv'
eva.save_results(save_dir, results)
```
